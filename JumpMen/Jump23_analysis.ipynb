## Jumpmen Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
    
```

```python
df = pd.read_csv('analyze_me.csv')
df.head()
```

```python
df.info()
```

## Null Values

```python
print('All null values are in yellow')
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
#Getting the fraction of the missing values for each variable

null_records = round(df.isnull().sum()*100/df.isnull().count(),2)
print(null_records)
```

```python
sorted(null_records, reverse=True)
```

```python
null_records.plot(kind='bar', color='Red')
plt.title('Percentage of null values in all columns')
```

## Jumpmen Data

```python
deliveries = pd.read_csv('deliveries_per_jumpman.csv')
deliveries.head()
```

```python
plt.figure(figsize=(10,10))
sns.countplot(x="AVG_deliveries", data=deliveries)
```

```python
print("Statistical Information")
deliveries['AVG_deliveries'].describe()
```

```python
print("Boxplot of delivered orders per Jumpmen")
plt.figure(figsize=(8,8))
sns.boxplot(deliveries['AVG_deliveries'])
```

```python
print('Distribution of Delivered orders per Jumpmen')
plt.figure(figsize=(8,8))
sns.distplot(deliveries['AVG_deliveries'])
```

## Correlation

```python
correlation = df.corr()
plt.figure(figsize=(25,25))
sns.heatmap(correlation, square=True, annot=True, linewidths=.5)
plt.title("Correlation Matrix")
```
